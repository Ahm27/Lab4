### loc vs. iloc
* **loc** stands for Location - it finds the data by its labels (names of rows and columns).
* loc[index label, column label]
* When passing a slice, loc is **inclusive** meaning the last value in the range will also be returned.
---
* **iloc** stands for Integer Location - it does not care about labels, only numerical positions.
* iloc[index number, column number]
* When passing a slice, iloc is **exclusive** meaning the last number is not included.

```python
print(df.loc[0:3, ['Name', 'Score']]) # loc[rows, columns] - inclusive so row 3 is included
print()
print(df.loc[0]) # returns the entire first row
print()
print(df.loc[0, 'Name']) # returns single value at row 0, column 'Name'
```

```python
print(df)
print()
print(df.iloc[0])    # returns first row - same as df.iloc[0, :]
print()
print(df.iloc[5, 2]) # returns value at row index 5, column index 2
```

### Filters
Filters allow us to extract rows from a DataFrame based on specific conditions.
* Use **&** for AND conditions (both must be true).
* Use **|** for OR conditions (at least one must be true).

```python
print(df)
print('\n=============================\n')

# Single condition filter - Score greater than 90
print('Filtered (Score > 90):\n')
print(df[df['Score'] > 90])

print('\n===================================')
# Multiple conditions - Score between 80 and 90
print('Filtered (80 < Score < 90):\n')
print(df[(df['Score'] > 80) & (df['Score'] < 90)])
```

### Handling Missing Values
Machine learning models cannot be trained on data that contains missing (NaN) values.

**Two approaches:**
1. **dropna()** - Remove rows that contain missing values. Best when the number of missing rows is small.
2. **fillna()** - Fill missing values with a replacement (e.g. mean). Best when there are many missing values and removing them would lose too much data.

**inplace=True** means modify the original DataFrame directly without creating a copy.

```python
# Check how many missing values exist in each column
df = pd.read_csv('data.csv') 
print('Missing values per column:')
print(df.isnull().sum())


# Option 1: Drop rows that have any missing values
print('\n=== After dropna() ===')
df_cleaned = df.dropna()
print(df_cleaned)

# Option 2: Fill missing values instead of removing them
print('\n=== After fillna() ===')
df_filled = df.copy()
df_filled['Score'].fillna(df_filled['Score'].mean(), inplace=True) # fill numbers with mean
df_filled['Name'].fillna('Unknown', inplace=True)                  # fill text with placeholder
print(df_filled)
```

### Removing Duplicates
Duplicate rows are rows that appear more than once in the dataset.
They can bias the model by giving certain data points more weight than they should have.

* **duplicated().sum()** - counts how many duplicate rows exist.
* **drop_duplicates()** - removes all duplicate rows, keeping only the first occurrence.

```python
# Count duplicate rows
print('Number of duplicate rows:', df.duplicated().sum())
print('Duplicate rows:\n', df[df.duplicated()])

# Remove all duplicate rows
# df_no_duplicates = df.drop_duplicates()
# print('\nAfter removing duplicates:')
# print(df_no_duplicates)

duplicates = df[df.duplicated(subset=['Name'],keep=False)]
print('\nDuplicate rows based on Name and Score:\n')
print(duplicates)   
```

### Pivot Table
A Pivot Table takes flat column-wise data and reorganizes it into a 2D summary table - just like Excel pivot tables.

Parameters:
* **index** - the column that becomes the rows
* **columns** - the column that becomes the column headers
* **values** - the column whose values will be computed
* **aggfunc** - the aggregation function to apply (sum, mean, count...)

```python
# Sample data with categories
data2 = {
    'Name':    ['Alice', 'Bob', 'Alice', 'Bob', 'Alice'],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math'],
    'Score':   [85, 90, 78, 88, 92]
}
df2 = pd.DataFrame(data2)
print('Original Data:')
print(df2)

# Pivot table: total score per person per subject
print('\n=== Pivot Table (sum of scores per Name per Subject) ===')
pivot = df2.pivot_table(index=['Name'], columns=['Subject'], values=['Score'], aggfunc=np.sum)
print(pivot)
```

### Encoding Categorical Data
Machine learning models only work with numbers. Text categories like 'Honda' or 'Blue' must be converted to numerical values.

**Two approaches:**

1. **One-Hot Encoding (get_dummies)** - Creates a new binary column for each category (0 or 1). Use this when there is NO order between categories (e.g. colors, car brands).

2. **Label Encoding** - Assigns a unique integer to each category. Use this when there IS an order between categories (e.g. Small=0, Medium=1, Large=2).

```python
df_cat = pd.DataFrame({
    'Name':  ['Alice', 'Bob', 'Charlie', 'Alice'],
    'Grade': ['A', 'B', 'A', 'C'],
    'Score': [95, 80, 90, 88],
    
})
print('Original:')
print(df_cat)

# One-Hot Encoding - creates Grade_A, Grade_B, Grade_C columns with 0 or 1
print('\n=== One-Hot Encoding (get_dummies) ===')
df_onehot = pd.get_dummies(df_cat, columns=['Grade']).astype({
    'Grade_A': int, 'Grade_B': int, 'Grade_C': int
})
print(df_onehot)

# Label Encoding - assigns a number to each unique category
print('\n=== Label Encoding ===')
df_label = df_cat.copy()
label_encoder = LabelEncoder()
df_label['Name_Encoded'] = label_encoder.fit_transform(df_label['Name'])
print(df_label)
```

### Handling Outliers - IQR Method
An outlier is a value that is far away from the rest of the data and can negatively affect the model.

Example: Ages = [20, 22, 21, 23, 300] -> 300 is an outlier.

**IQR (Interquartile Range) Method:**

2,3,4,5,6,7,8,9,40
* Q1 = 25th percentile (lower quarter of data)
* Q3 = 75th percentile (upper quarter of data)
* IQR = Q3 - Q1
* Lower Bound = Q1 - 1.5 * IQR
* Upper Bound = Q3 + 1.5 * IQR
* Any value outside these bounds is considered an outlier.

**Two ways to handle outliers:**
1. **Remove** - drop the rows containing outliers.
2. **Cap (clip)** - replace outliers with the boundary values instead of removing them.

```python
df_out = pd.DataFrame({
    'Score': [85, 90, 78, 95, 88, 73, 82, 89, 94, 10, 300]
})
print('Original Data:')
print(df_out)

# Calculate IQR bounds
Q1 = df_out['Score'].quantile(0.25)
Q3 = df_out['Score'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f'Q1={Q1}, Q3={Q3}, IQR={IQR}')
print(f'Lower Bound={lower_bound}, Upper Bound={upper_bound}')

# Identify outliers
outliers = df_out[(df_out['Score'] < lower_bound) | (df_out['Score'] > upper_bound)]
print('\nOutliers found:')
print(outliers)

# Option 1: Remove outlier rows completely
df_removed = df_out[df_out['Score'].between(lower_bound, upper_bound)]
print('\n=== After Removing Outliers ===')
print(df_removed)

# Option 2: Cap outliers at the boundary values
df_capped = df_out.copy()
df_capped['Score'] = df_capped['Score'].clip(lower=lower_bound, upper=upper_bound)
print('\n=== After Capping Outliers ===')
print(df_capped)
```

### Scaling & Normalization
When features have very different scales, the model becomes biased toward features with larger numbers.

Example: Score ranges from 0-100 but Salary ranges from 0-100,000. The model will focus more on Salary just because its numbers are bigger.

**Min-Max Normalization** rescales all values to be between 0 and 1:

new_value = (value - min) / (max - min)

* The smallest value in the column becomes 0
* The largest value in the column becomes 1
* Everything else falls in between

```python
df_scale = pd.DataFrame({
    'Name':  ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 92, 78, 95],
    'Age':   [20, 25, 22, 30]
})
print('Before Normalization:')
print(df_scale)

# Apply Min-Max Normalization to the Score column
df_scale['Score_Normalized'] = (
    (df_scale['Score'] - df_scale['Score'].min()) /
    (df_scale['Score'].max() - df_scale['Score'].min())
)

print('\nAfter Normalization (Score values now between 0 and 1):')
print(df_scale)
```

### Handling Datetime Data
Pandas reads date columns as plain strings by default. We need to convert them to datetime objects so we can extract useful information like day, month, year, and day of the week.

* **pd.to_datetime()** - converts a string column into a proper datetime column.
* Once converted, use **.dt** accessor to extract date parts.

```python
df_date = pd.DataFrame({
    'Name':      ['Alice', 'Bob', 'Charlie', 'David'],
    'Score':     [85, 92, 78, 95],
    'Exam_Date': ['2024-01-15', '2024-03-22', '2023-11-05', '2024-06-10']
})

# Before conversion - the date is just a string
print('Before conversion - type is:', type(df_date['Exam_Date'].iloc[0]))

# Convert the column to datetime
df_date['Exam_Date'] = pd.to_datetime(df_date['Exam_Date'])

# After conversion - now it is a proper Timestamp
print('After conversion  - type is:', type(df_date['Exam_Date'].iloc[0]))

# Extract useful date parts using .dt accessor
df_date['day']         = df_date['Exam_Date'].dt.day
df_date['month']       = df_date['Exam_Date'].dt.month
df_date['year']        = df_date['Exam_Date'].dt.year
df_date['day_of_week'] = df_date['Exam_Date'].dt.day_name()

print('\nFinal DataFrame:')
print(df_date)
```
